```python
folder_name = 'bestofrt_posters'
if not os.path.exists(folder_name):
    os.mkdir(folder_name)
```

```python
df_list = []
df_errors = {}
for title in title_list:
    try:
        ranking = title_list.index(title) + 1
        page = wptools.page(title, silent=True)  
        page.get()
        images = page.data['image']
        img_url = page.data['image'][0]['url']
        fname = img_url.split('/')[-1]
#         Since i already ran, hence using this
        if os.path.exists(folder_name+'/'+fname):
            df_list.append({'rank': ranking, 'title': title, 'image_url': img_url})
            print(f'{title} was already a file, skipping')
            continue
        r = re.get(img_url)
        print(f'capturing for: {fname}')
        with Image.open(BytesIO(r.content)) as im:
            im.save(f'{folder_name}/{fname}')
            df_list.append({'rank': ranking, 'title': title, 'image_url': img_url})
            print(f'executed despite failure for title:{title}')
    except:
        print('something is not right')
        img_list = []
        for image in images:
            img_list.append(image['url'])
        print(f'URLs for errored titles {img_list}')
        df_errors[str(ranking)+"_"+title] = img_list
```

```python
df_images = pd.DataFrame(df_list)
```

```python
# Iterating over failed images
for title in df_errors:
    try:
        for url in df_errors[title]:
            print(url)
            r = re.get(url)
            fname = '_'.join(title.split('_')[1:])
            rank = title.split('_')[0]
            with Image.open(BytesIO(r.content)) as im:
                im.save(f'{folder_name}/{fname}.jpg')
            if os.path.exists(f'{folder_name}/{fname}.jpg'):
                print(f'{title} is now found')
                df_list.append({'rank': rank, 'title': fname, 'image_url': url})
                break
    except:
        print(f'{title} not a valid downloadable image')
        continue
```
