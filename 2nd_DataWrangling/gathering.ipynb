```python
df_list = []
for file in os.listdir('rt_html/'):
    with open(f'rt_html/{file}', mode='r') as html:
        soup = BeautifulSoup(html, 'lxml')
        title = soup.find('title').contents[0][:-len(' - Rotten Tomatoes')]
        score = soup.find('div', class_='audience-score meter').find('span').contents[0][:-1]
        num_rev = soup.find('div', class_="audience-info hidden-xs superPageFontColor").find_all('div')[1].contents[2].replace(',', '').strip()
        df_list.append({'title': title, 'percent':score, 'num_reviews':num_rev})
        
        
```

```python
df_rt = pd.DataFrame(df_list)
```
